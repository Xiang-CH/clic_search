```python
import pandas as pd
```

```python
eng = pd.read_excel('data/CLIC_Content_List.xlsx')
tc = pd.read_excel('data/CLIC_Content_TC_20230809.xlsx')
```

```python
eng.head(5)
```

```python
tc.head(5)
```

```python
eng.shape
```

```python
tc.shape
```

```python
tcExtra = pd.DataFrame(columns=['nid','title','path','content'])
tc['topic'] = ""
for index, row in tc.iterrows():
    path =  row['path'].replace('/cn/', '/en/')
    print("hk path: " + path)

    topic = eng.loc[eng['full_path'] == path, 'topic']
    if topic.empty:
        print("English page not found")
        # topic = path.split("/topics/")[1].split("/")[0]
        tcExtra = tcExtra.append(row, ignore_index=True)
        tc.drop(index, inplace=True)
    else:
        tc.at[index, 'topic'] = topic.to_string(index=False)
        tc.at[index, 'path'] = path.replace('/en/', '/zh/')
    
    print(topic)
```

```python
tc.to_excel('data/tc_with_topic.xlsx')
tcExtra.to_excel('data/tc_extra.xlsx')
```
